```python
regex = r'^(\d{4})'

# The \d represents any digit, and {4} repeats this rule four times.
# The ^ character matches the start of a string
# the parentheses denote a capturing group, which signals to pandas that we want to extract that part of the regex.
 # (We want ^ to avoid cases where [ starts off the string.)

# The regular expression above is meant to find any four digits at the beginning of a string
# 'r' is a raw string (meaning that a backslash is no longer an escape character),
# which is standard practice with regular expressions.
```

```python
extr = df['Date of Publication'].str.extract(r'^(\d{4})', expand=False)
extr.head()
```

```python
df['Date of Publication'] = pd.to_numeric(extr)
df['Date of Publication'].dtype

# changes of 'Date of Publication' column dtype to numeric
# from object
# This results in about one in every ten values being missing,
```

**Combining str Methods with NumPy to Clean Columns**

```python
# we can combine pandas str methods with NumPy’s np.where function,
# which is basically a vectorized form of Excel’s IF() macro
# the syntax is: np.where(condition, then, else)

# 'condition' is either an array-like object or a Boolean mask
# 'then' is the value to be used if 'condition' evaluates to True
# 'else' is the value to be used otherwise
# .where() takes each element in the object used for condition,
# checks whether that particular element evaluates to True in the context of the condition,
# and returns an ndarray containing then or else, depending on which applies.

# It can be nested into a compound if-then statement, allowing us to compute values based on multiple conditions:
# >>> np.where(condition1, x1,
#        np.where(condition2, x2,
#            np.where(condition3, x3, ...)))

# we'll use these two functions to clean Place of Publication since this column has string objects
```

```python
df['Place of Publication'].head(10)
```

```python
df.loc[4157862]
```

```python
df.loc[4159587]
```

```python
# The two books above were published in the same place, but one has hyphens in the name of the place while the other does not.
# To clean this column in one sweep, we can use str.contains() to get a Boolean mask.
```

```python
pub = df['Place of Publication']
london = pub.str.contains('London')
london[:5]
```

```python
oxford = pub.str.contains('Oxford')
oxford[:5]
```

```python
df['Place of Publication'] = np.where(london, 'London',
                                      np.where(oxford, 'Oxford',
                                               pub.str.replace('-', ' ')))
```

**Cleaning the Entire Dataset Using the applymap Function**

```python
# creating a list of (state, city) tuples and wrapping that list in a DataFrame
# file source: https://github.com/realpython/python-data-cleaning/blob/master/Datasets/university_towns.txt
# upload the file to GC

university_towns = []
with open('university_towns.txt') as file:
  for line in file:
    if '[edit]' in line:
      # Remember this `state` until the next is found
      state = line
    else:
      # Otherwise, we have a city; keep `state` as last-seen
      university_towns.append((state, line))

university_towns[:5]
```

```python
# set the columns as “State” and “RegionName”
# pandas will take each element in the list and set State to the left value and RegionName to the right value

towns_df = pd.DataFrame(university_towns,
                        columns=['State', 'RegionName'])

towns_df.head()
```

```python
def get_citystate(item):
  if ' (' in item:
    return item[:item.find(' (')]
    # [:item.find()] is a string slicing to extract string before ' ('
    # [start:end], empty start means starts from index 0 of str
    # the end is ' (' and will not be included in the returned output
  elif '[' in item:
    return item[:item.find('[')]
  else:
    return item
```

```python
# pandas’ .map() only takes one parameter,
# which is the function (callable) that should be applied to each element

towns_df =  towns_df.map(get_citystate)

# The applymap() method took each element from the DataFrame,
# passed it to the function, and the original value was replaced by the returned value.
```

```python
towns_df.head()
```

**Renaming Columns and Skipping Rows**

```python
# rename columns and skip certain rows so that we can drill down to necessary information with correct and sensible labels
# file source: https://github.com/realpython/python-data-cleaning/blob/master/Datasets/olympics.csv
# upload the file to GC

olympics_df = pd.read_csv('olympics.csv')
olympics_df.head()
```

```python
olympics_df = pd.read_csv('olympics.csv', header=1)
# can use index_col=0 instead of header=1 to get rid of the "Unnamed: 0" column
# and making the first column the first column + get rid of the index number on the left
olympics_df.head()
```

```python
# defining a dictionary that maps current column names (as keys) to more usable ones (the dictionary’s values)

new_names =  {'Unnamed: 0': 'Country',
              '? Summer': 'Summer Olympics',
              '01 !': 'Gold',
              '02 !': 'Silver',
              '03 !': 'Bronze',
              '? Winter': 'Winter Olympics',
              '01 !.1': 'Gold.1',
              '02 !.1': 'Silver.1',
              '03 !.1': 'Bronze.1',
              '? Games': '# Games',
              '01 !.2': 'Gold.2',
              '02 !.2': 'Silver.2',
              '03 !.2': 'Bronze.2'}
```

```python
# To rename the columns, we will make use of a DataFrame’s rename() method,
# which allows you to relabel an axis based on a mapping (in this case, a dict)
# call the rename() function on our object

olympics_df.rename(columns=new_names, inplace=True)

# Setting inplace to True specifies that our changes be made directly to the object
```
